```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd

infile = '/Users/timseifert/multi-object-tracker/examples/example_notebooks/bounding_box_collection_cars.csv'



def plotLine(p1, p2):
    plt.plot( [p1[0], p2[0]], [p1[1], p2[1]] )

COLS = ['id','x','y']
df = pd.DataFrame(columns=COLS)


x =0
x2=0
y=0
y2=0
x_final=[]
y_final = []
id_point =[]


with open(infile, 'r') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        try:
            coords = row[1]
            coords = coords.strip("[]")
            arr = coords.split(" ")
            while '' in arr:
                arr.remove('')
            x = float(arr[0])
            x2 = float(arr[1])
            y = (float(arr[2]))
            y2 = float(arr[3])

            x_final.append(((x+x2)))
            #we must invert y coordinates
            y_final.append(((y+y2)))
            print("x is" + str(float(x)))
            print("y is" + str(float(y)))
            id_point.append(row[0])
            # take coordinates and output to a csv file
            new_entry = []
            new_entry.append(row[0])
            new_entry.append((x+y2))
            new_entry.append((y+x2))
        
            data_frame = pd.DataFrame([new_entry], columns=COLS)
            df = df.append(data_frame, ignore_index=True)  
            df.to_csv('id_and_coordinates.csv', columns=COLS,index=False) 
        except:
            continue
     
# Fill a dictionary with the points
points = {} # Create empty dictionary
for idx, point in enumerate(zip(x_final,y_final)):
    points[id_point[idx]] = point 


#'/Users/timseifert/multi-object-tracker/examples/assets/ppl_walking.png'
im = plt.imread('/Users/timseifert/multi-object-tracker/examples/assets/cars.png')
plt.figure(figsize=(20,20))
implot = plt.imshow(im)
plt.scatter(x_final, y_final)

for key in points:
    plt.annotate(key, xy=points[key])
for id_val in id_point:
    print(points[id_val])


plt.gca()
plt.show()
```
